## Importing packages

```python
import os, sys

sys.path.insert(0, os.path.dirname(os.path.abspath('..')))
from mass_automation.formula import Formula
from mass_automation.formula.plot import plot_compare
from mass_automation.experiment import Experiment
```

Compound presence verification can be applied on tea samples to detect ions. Let's see few examples  

<b> !! The path below is just an example, specify path, where you saved folder with .mzXML spectra of tea extracts by yourself !! </b>

```python
path = 'D:\\mass_spectra\\TEA_dataset'
```

```python
for spectrum, formula, name in zip(['Tea_15_33_01_173.mzXML',
                                    'Tea_3_20_01_134.mzXML',
                                    'Tea_4_21_01_139.mzXML',
                                    'Tea_6_23_01_144.mzXML'],
                                   
                                   ['C22H18O11K',
                                    'C27H30O15K',
                                    'C21H20O10K',
                                    'C21H20O11K'],
                                   
                                   ['(-)-Epigallocatechin gallate_(-)-Gallocatechin gallate',
                                    'Kaempferol-3-O-rutinoside',
                                    'Vitexin_K',
                                    'Quercitrin_kaempferol-3-glucoside']):
    
    exp = Experiment(os.path.join(path, spectrum))  # Create Experiment object   
    #exp = Experiment(f'D:\\mass_spectra\\TEA_dataset\\{spectrum}')
    spectrum = exp.summarize(4, 9)   # Summarize some spectra in Experiment to create total spectrum
    plot_compare(spectrum,  # Spectrum, where we try to find ion
                 Formula(formula),  # Formula object, containing ion name
                 cal_error=0.01   # Calibration error accounting
                )
```

Now, you can use it with any ion you want!
