### Cluster maps

are important visualization techniques in biology to find correlations between genes. This approach can be applied to mass spectra. Can't wait to show you a real example with the tea spectra

First, let's load spectra vectors dataset, which we made in *mass_spectra_vectorization* notebook

```python
spec_path = os.path.join(os.path.dirname(os.path.abspath('..')), 'data', 'plot_pca_files', 'spec_vecs_dictionary.pkl')
```

```python
with open(spec_path, 'rb') as f:
    spec_vecs = pkl.load(f)
```

Now, let's choose required spectra from spectra vectors dictionary and fill in information about their class in *class_decoder* 

```python
required_keys = ['Tea_14_31_01_167.mzXML', 'Tea_10_27_01_156.mzXML', 'Tea_17_34_01_177.mzXML', 'Tea_5_22_01_140.mzXML', 
                 'Tea_6_23_01_143.mzXML', 'Tea_9_26_01_152.mzXML', 'Tea_11_28_01_159.mzXML', 'Tea_2_19_01_131.mzXML', 'Tea_4_21_01_137.mzXML']
class_decoder = ['Green', 'Green', 'Green', 'Pu-erh', 'Black', 'Black', 'Black', 'Massala', 'Cham']
```

```python
agglosamples(spec_vecs,
            required_keys,
            class_decoder,
            show=True)
```

We created sample cluster maps, we see black teas located with other black teas, so do green teas. Chamomile tea is closer to green tea than black one. Pu-erh is closer to black teas
